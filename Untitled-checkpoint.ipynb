## Cartoonify an Image with OpenCV

#### Cartoonify an Image with OpenCV

```python
#import librabries
import cv2 #for imageprocessing
import easygui #to open filebox
import numpy as np
import imageio  #to read image stored at particular path
import sys
import matplotlib.pyplot as plt
```

```python
ImagePath=('Picture.jpg')
```

```python
img=cv2.imread('Picture.jpg')
originalmage=cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    
if originalmage is None:
    print("can not find image.....")
    sys.exit()
    
ReSized1=cv2.resize(originalmage,(960,940))
plt.imshow(ReSized1, cmap='gray')
```

```python
#convert into grey
grayScaleImage = cv2.cvtColor(originalmage, cv2.COLOR_BGR2GRAY)
ReSized2 = cv2.resize(grayScaleImage, (1160, 1040))
plt.imshow(ReSized2, cmap='gray')
```

```python
smoothGrayScale=cv2.medianBlur(grayScaleImage,5)
ReSized3 = cv2.resize(smoothGrayScale,(1160,1040))
plt.imshow(ReSized3,cmap='gray')
```

```python
getEdge = cv2.adaptiveThreshold(smoothGrayScale,500 ,cv2.ADAPTIVE_THRESH_MEAN_C,cv2.THRESH_BINARY,9,9)
ReSized4=cv2.resize(getEdge,(1160,1040))
plt.imshow(ReSized4,cmap=('gray'))
```

```python
colorImage = cv2.bilateralFilter(originalmage,9,300,300)
ReSized5 = cv2.resize(colorImage,(1160,1040))
plt.imshow(ReSized5,cmap=('gray'))
```

```python
cartoonImage = cv2.bitwise_and(colorImage, colorImage , mask=getEdge)
ReSized6 =cv2.resize(cartoonImage,(1500,1200))
plt.imshow(ReSized6, cmap=('gray'))
```

```python
images=[ReSized1, ReSized2, ReSized3, ReSized4, ReSized5, ReSized6]
fig, axes = plt.subplots(3,2, figsize=(15,15), subplot_kw={'xticks':[], 'yticks':[]}, gridspec_kw=dict(hspace=0.1, wspace=0.1))
for i, ax in enumerate(axes.flat):
    ax.imshow(images[i], cmap='gray')
plt.show()
```
